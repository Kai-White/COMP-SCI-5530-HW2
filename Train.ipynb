```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Kai-White/COMP-SCI-5530-HW2/refs/heads/main/Source/train.csv')
```

```python
df.head(10)
```

```python
df.isna().mean()
```

86% of New_Price entries are blank. It would not be smart to replace these values with mean, median, or mode because it is the target variable. If we want to be focused on New_Price, we don't want our data to be based on created values.

Once null New_Price entries are removed, there are no other variables missing except for a miniscule amount of missing mileage entries. We can afford to drop them.  


```python
df = df.dropna(subset=['New_Price'])
df = df.dropna(subset=['Mileage'])
df.isna().mean()
```

```python
df.head()
```

Removing units

```python
df['Mileage']=df['Mileage'].astype(str)
df['Mileage'] = (
    df['Mileage']
    .str.replace(',', '', regex=False)
    .str.extract(r'(\d+\.?\d*)', expand=False)
)
df['Engine']=df['Engine'].astype(str)
df['Engine'] = (
    df['Engine']
    .str.replace(',', '', regex=False)
    .str.extract(r'(\d+\.?\d*)', expand=False)
)
df['New_Price']=df['New_Price'].astype(str)
df['New_Price'] = (
    df['New_Price']
    .str.replace(',', '', regex=False)
    .str.extract(r'(\d+\.?\d*)', expand=False)
)
df['Power']=df['Power'].astype(str)
df['Power'] = (
    df['Power']
    .str.replace(',', '', regex=False)
    .str.extract(r'(\d+\.?\d*)', expand=False)
)
df.head()
```

```python
one_hot = pd.get_dummies(df['Fuel_Type'].apply(pd.Series).stack()).groupby(level=0).sum()
df_one_hot = df.drop('Fuel_Type', axis=1).join(one_hot)
one_hot_transmission = pd.get_dummies(df_one_hot['Transmission'].apply(pd.Series).stack()).groupby(level=0).sum()
new_df = df_one_hot.drop('Transmission', axis=1).join(one_hot_transmission)
new_df
```

Adding car brand column

```python
df["Brand"]= df["Name"].str.split().str.get(0)
df
```
